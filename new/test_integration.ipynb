```python
import numpy as np

#define starting points
r0=1e14 #cm
rho0 = 1e-10 #g cm^-3

#make a test array of radii to consider:
test_radii = np.logspace(np.log10(1e14),np.log10(2e15),num=100)

#figure out the density at each of these radii:
test_rho = (r0/test_radii)**2. * rho0

#Make the integrad for the integral that we want to do over the volume:
test_integrand = 4 * np.pi * test_radii**2. * test_rho

#Loop over each value in this test array to calculate the integral between r0 and that radius:
for i in range(len(test_radii)-1):
    mass_within_g = np.trapz(test_integrand[:i+2],test_radii[:i+2])
    mass_within_Msun = mass_within_g/1.989e33
    print(len(test_radii[:i+2]),test_radii[i+1], test_rho[i+1],mass_within_g,mass_within_Msun)
```
