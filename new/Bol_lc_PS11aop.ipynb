```python
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
file='bol1.csv'
data = ascii.read(file,format='csv')
t = data['t']

lum = data['lum']
lumerror =  data['lum-error']
yerr =  lumerror
plt.errorbar(t,lum,yerr,fmt = 'k.',label = 'Bol LC PS11aop')
plt.plot(t,lum,'k')
#plt.plot(r1,enuc2,'r',label = '0.5 solar mass')
#plt.scatter(logTeff,logL)
#plt.xscale('log')
plt.yscale('log')
plt.xlabel('t (days)',fontsize=20)
plt.ylabel('Luminosity',fontsize=20)
plt.legend(fontsize=20)
#plt.xlim(5.0,3.0)
plt.ylim(5.5e42,1e44)
#plt.show()
plt.savefig('bol1.pdf')
```
