```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
```

## Read in Results:

```python
file='Results.csv'
data = ascii.read(file,format='csv')
```

## HR Diagram:

```python
logTeff = np.log10(data['Teff'])
logL = data['lg_L']

plt.figure()

#Model data:
plt.plot(logTeff,logL)
plt.scatter(logTeff,logL)

#Homology relation #1:
x = np.linspace(3,6,100)

slope = 5# This should be the exponent of the homology relationship
offset = -17 #You can set this to whatever to move the line close to your model data.
y = slope*x+offset # This is just linear because we are ploting log-log

plt.plot(x,y,color='r')

plt.xlabel('Log(Teff)')
plt.ylabel('Log(L/Lsun)')
plt.xlim(6.0,3.0)
#plt.ylim(-2,7)
plt.show()
```
