```python
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
file='bol1.csv'
data = ascii.read(file,format='csv')
t = data['t']

lum = data['lum']
lumerror =  data['lum-error']
yerr =  lumerror
plt.errorbar(t,lum,yerr,fmt = 'k.',label = 'Bolometric light curve for PS11aop')

#plt.plot(r1,enuc2,'r',label = '0.5 solar mass')
#plt.scatter(logTeff,logL)
plt.xlabel('t (days)')
plt.ylabel('Luminosity')
plt.legend()
#plt.xlim(5.0,3.0)
#plt.ylim(-10,10)
#plt.show()
plt.savefig('bol1.pdf')
```
